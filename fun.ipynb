```python
%matplotlib inline

import matplotlib.pyplot as plt
import math
import numpy
import numpy.linalg

def r(x):
    return r0 + 5*x

r0 = 5
f = 0
pi = math.pi
xs = []
ys = []

while f <30*pi:
    xs.append(r(f)*math.cos(f))
    ys.append(r(f)*math.sin(f))
    f +=pi/180
    
plt.plot(xs,ys)

plt.show()
```

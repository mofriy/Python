```python
import random
#у меня данных с лаб. работ нету, поэтому так
f = open('data.txt', 'w')
for i in range(1000):
    t = 0
    for k in range(50):
        t += random.randint(0,1)
    f.write(str(t) + '\n')
```

```python
%matplotlib inline

import pandas
import matplotlib.pyplot as plt

f = open('data.txt', 'r')
experiments = []
for num in f:
    experiments.append(float(num))
#plt.plot(list(range(len(experiments))), experiments)
#plt.title('My DATA 3000')
#plt.show()

df = pandas.DataFrame(data={
    'experiments': experiments
})

df.to_csv("experiments.csv")

df1 = pandas.read_csv("experiments.csv")
df1['experiments'].plot()
```
